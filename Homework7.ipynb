**The G dwarf problem**, simply stated, is that astronomers observe a lower fraction of metal-poor G dwarfs in the solar neighborhood than would be expected from closed-box models.

How might this "infall" model address the G dwarf problem? As seen in the above plots, and assuming a ~constant star-forming rate, the system spends relatively little time in its low-metallicity state before advancing to higher metallicities (note that the points are not evenly spaced along the x-axis, but bunch up towards low $\mu$.) Thus, the infall "stretches out" the metallicity history, making it harder to observe low-metallicity G dwarfs.

Yes.** We know that low-mass galaxies have high rates of outflow relative to their star formation rates, which is the same as a high $\eta$ here. We have just seen above that high $\eta$ corresponds to low metallicity. Thus we can explain the metallicity-mass relation as a signature of outflows (parameterized by $\eta$) removing most of the metals in a low-mass galaxy before they are allowed to be accumulated in stars.
